```python
import numpy as np
```

```python
class Vector:
    __values = []
    
    def __init__(self, values):
        if (type(values) == list):
            self.__values = values
        else:
            raise Exception("Вектор должен быть представлен списком значений.")
    
    def __add__(a, b):
        if len(a.__values) != len(b.__values):
            raise Exception("Вектора должны быть одинаковой размерности.")
        
        return Vector([a.__values[i] + b.__values[i] for i in range(len(a.__values))])
    
    @property
    def length(self):
        return np.sqrt(sum([val**2 for val in self.__values]))
    
    def __str__(self):
        return str(self.__values)
```

```python
a = Vector([10, 10, 10])
b = Vector([0, 0, -10])
c = a + b
print(c)
```

```python
print(a.length)
print(b.length)
print(c.length)
```

```python
import matplotlib.pyplot as plt
import decimal
import numpy as np
```

Окружность

```python
def circle(x, r, pos=True):
    y = np.sqrt(r**2 - x**2)
    return y if pos else -y


r = 300
n_points = 10000
points = np.linspace(-r, r, n_points)
y1 = [circle(x, r) for x in points]
y2 = [circle(x, r, False) for x in points]
```

```python
plt.figure(figsize = (5, 5))    
plt.plot(points, y1, 'blue')  
plt.plot(points, y2, 'blue')
plt.axis('scaled')
plt.show()
```

Эллипс

```python
def ellipse(x, a, b, pos=True):
    y = np.sqrt(b**2 - b**2 * x**2 / a**2)
    return y if pos else -y


a, b = 50, 20
n_points = 10000
points = np.linspace(-a, a, n_points)
y1 = [ellipse(x, a, b) for x in points]
y2 = [ellipse(x, a, b, False) for x in points]
```

```python
plt.plot(points, y1, 'blue')  
plt.plot(points, y2, 'blue')
plt.show()
```

Решите систему уравнений и неравенств:
y = x2 – 1
exp(x) + x∙(1 – y)  - 1 > 0


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
# Решите систему уравнений:
# y = x**2 – 1
# exp(x) + x∙(1 – y) = 1
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) - x*y +x -1)

# построим графики
x = np.linspace(-2,3,201)
plt.plot(x,(np.exp(x) +x - 1)/x) 
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.5)
plt.grid(True)
plt.show()

# решим с-му численным методом
#1
xx = -1.5
yy = 1.5
print("Решение с точкой приближения: (",xx,";",yy,")")
x1, y1 = fsolve(equations,(xx,yy))
print ("x1=",x1,"; y1=",y1)
#2
xx = 2.5
yy = 5.5
print("Решение с точкой приближения: (",xx,";",yy,")")
x2, y2 = fsolve(equations,(xx,yy))
print ("x2=",x2,"; y2=",y2)

print()
# Решите систему уравнений и неравенств:
# y = x**2 – 1
# exp(x) + x∙(1 – y) > 1

def ner(x,y):
    return np.exp(x)+x*(1-y)-1
print ("Выше мы нашли корни системы уравнений. Проверим найденные корни, подставив их в систему с уравнением и неравенством.")
print("y = x**2 – 1")
print ("exp(x) + x∙(1 – y) -1 > 0")
print("Решением системы будут корни, при котором выполняется неравенство (>0 ). Подставим и посчитаем:")
print("x1,y1:",ner(x1,y1))
print("x2,y2:",ner(x2,y2))
```
